The selected approach based on the analysis is to drop the null values for certain columns in the dataset. This decision was made based on the following observations:

1. The columns with the highest percentage of null values are "Developer Website" and "Privacy Policy", which are not essential for the analysis. Therefore, we can drop these columns.

3. Columns such as "Released", "Rating", and "Rating Count" are important, and it is better to fill their null values rather than dropping them.

Therefore, the selected approach is to drop the null values of the selected columns to ensure the integrity and accuracy of the analysis.

# Analysis:
After analyzing the content rating column, we have identified that there are multiple categories in this column, including:

* Everyone
* Teen
* Mature 17+
* Everyone 10+
* Unrated
* Adults only 18+

To simplify these categories and make them more understandable, we can group them into 3 broader categories:

* Everyone
* Teen
* Adults (which includes Mature 17+ and Adults only 18+)
* Unrated apps can be categorized as Everyone as there is no specific age limit mentioned.

By doing this, we can have a more clear and concise understanding of the content rating of the apps.


```python
df["Content Rating"] = df["Content Rating"].str.split("+").str[0]
```

```python
df["Content Rating"] = df["Content Rating"].replace("Unrated","Everyone")
df["Content Rating"] = df["Content Rating"].replace("Everyone 10","Teen")
df["Content Rating"] = df["Content Rating"].replace("Mature 17","Adults")
df["Content Rating"] = df["Content Rating"].replace("Adults only 18","Adults")
```

**Q1. What are the top 10 Categories that are installed from the Google Play Store?**

```python
df["Category"].unique()
```

```python
df["Category"].value_counts()
```

```python
top_category = df["Category"].value_counts().reset_index().rename(columns={"Category" : "Count", "index" : "Category"})
top_category
```

```python
category_installs = df.groupby(["Category"])[["Installs"]].sum()
category_installs
```

```python
top_category_installs = pd.merge(top_category, category_installs, on="Category")
top_10_category_installs = top_category_installs.head(10).sort_values(by=["Installs"], ascending=False).reset_index()
top_10_category_installs
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=60)
plt.xlabel("Category")
plt.ylabel("Number of Application")
plt.title("Top 10 Installed Category")
sns.barplot(x=top_10_category_installs["Category"], y=top_10_category_installs["Installs"])
```

**Answer:**
