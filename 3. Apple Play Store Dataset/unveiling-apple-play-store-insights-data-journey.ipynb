**1. let's first take a look at the "Content_Rating" column.** 

```python
df.Content_Rating.value_counts()
```

```python
df["Content_Rating"]=df["Content_Rating"].replace("4+","Children-Adults")
df["Content_Rating"]=df["Content_Rating"].replace("9+","Children-Adults")
df["Content_Rating"]=df["Content_Rating"].replace("12+","Teens-Aults")
df["Content_Rating"]=df["Content_Rating"].replace("17+","Only Adults")
df["Content_Rating"]=df["Content_Rating"].replace("Not yet rated","Everyone")
```

In this code, we are modifying the **"Content_Rating"** column of the DataFrame named "df". We are using the **.replace()** function to replace specific values in the **"Content_Rating"** column with more meaningful labels.

* **df["Content_Rating"]=df["Content_Rating"].replace("4+","Children-Adults"):** Here, we are replacing the value "4+" with "Children-Adults" in the "Content_Rating" column.
* **df["Content_Rating"]=df["Content_Rating"].replace("9+","Children-Adults"):** Similarly, we are replacing "9+" with "Children-Adults".
* **df["Content_Rating"]=df["Content_Rating"].replace("12+","Teens-Adults"):** Replacing "12+" with "Teens-Adults".
* **df["Content_Rating"]=df["Content_Rating"].replace("17+","Only Adults"):** Replacing "17+" with "Only Adults".
* **df["Content_Rating"]=df["Content_Rating"].replace("Not yet rated","Everyone"):** Replacing "Not yet rated" with "Everyone".

This code is transforming the "Content_Rating" values to more descriptive categories for better understanding and analysis.

**2. Let's take a look at the "Free" column.**

```python
df.Free.value_counts()
```

```python
df["Free"]=df["Free"].replace(True,"Free")
df["Free"]=df["Free"].replace(False,"Paid")
```

```python
df.rename(columns={'Free' : 'Type'}, inplace=True)
```

The code is modifying the **"Free"** column in the DataFrame. It replaces the values True with **"Free"** and False with **"Paid"**. Then, it renames the **"Free"** column to **"Type"**. This code is likely categorizing apps based on their pricing status as either **"Free"** or **"Paid"**.

**3. Let's take a look at the "Average_User_Rating" column.**

```python
df.Average_User_Rating.value_counts()
```

```python
df["Average_User_Rating_Round"] = df["Average_User_Rating"].round().astype(int)
```

The code is creating a new column called **"Average_User_Rating_Round"** in the DataFrame. It does this by rounding the values in the existing **"Average_User_Rating"** column to the nearest whole number and then converting those rounded values to integers. This likely aims to provide a simplified user rating for the apps based on their average user rating.

```python
df['Type_User_Rating'] = 'No Rating'
df.loc[(df['Average_User_Rating'] > 0) & (df['Average_User_Rating'] <= 1),'Type_User_Rating'] = 'Rating 0 to 1'
df.loc[(df['Average_User_Rating'] > 1) & (df['Average_User_Rating'] <= 2),'Type_User_Rating'] = 'Rating 1 to 2'
df.loc[(df['Average_User_Rating'] > 2) & (df['Average_User_Rating'] <= 3),'Type_User_Rating'] = 'Rating 2 to 3'
df.loc[(df['Average_User_Rating'] > 3) & (df['Average_User_Rating'] <= 4),'Type_User_Rating'] = 'Rating 3 to 4'
df.loc[(df['Average_User_Rating'] > 4) & (df['Average_User_Rating'] <= 5),'Type_User_Rating'] = 'Rating 4 to 5'
df['Type_User_Rating'].value_counts()
```

This code creates a new column named **'Type_User_Rating'** in the DataFrame 'df'. It assigns different rating categories based on the average user rating. The categories include 'No Rating' for apps without a rating and ranges from 'Rating 0 to 1' to 'Rating 4 to 5' for apps with different average user ratings. The code then counts the occurrences of each category in the **'Type_User_Rating'** column. This categorization allows better analysis and understanding of the distribution of app ratings.

```python
df['User_Rating'] = 'No Rating'
df.loc[(df['Average_User_Rating'] > 0) & (df['Average_User_Rating'] <= 5),'User_Rating'] = 'Yes Rating'
df['User_Rating'].value_counts()
```

This code creates a new column named **'User_Rating'** in the DataFrame 'df'. Initially, it assigns 'No Rating' to all rows in this column. Then, using the loc function, it checks if the **'Average_User_Rating'** is greater than 0 and less than or equal to 5. If this condition is met, the value in the **'User_Rating'** column is updated to 'Yes Rating'. Finally, the code counts the occurrences of each value in the **'User_Rating'** column. This step is intended to categorize apps as either having a user rating (within the range of 1 to 5) and not having a rating.

**4. Let's take a look at the "Reviews" column**

```python
df.Reviews.value_counts()
```

```python
df['Review_Category'] = 'No Reviews mention'
df.loc[(df['Reviews'] > 0) & (df['Reviews'] <= 1000.0),'Review_Category'] = 'Less than 1K Reviews'
df.loc[(df['Reviews'] > 1000.0) & (df['Reviews'] <= 10000.0),'Review_Category'] = 'Between 1K and 10K Reviews'
df.loc[(df['Reviews'] > 10000) & (df['Reviews'] <= 100000),'Review_Category'] = 'Between 10K and 100K Reviews'
df.loc[(df['Reviews'] > 100000) & (df['Reviews'] <= 500000),'Review_Category'] = 'Between 100K and 500K Reviews'
df.loc[(df['Reviews'] > 500000) & (df['Reviews'] <= 1000000),'Review_Category'] = 'Between 500K and 1M Reviews'
df.loc[(df['Reviews'] > 1000000),'Review_Category'] = 'More than 1M Reviews'
df['Review_Category'].value_counts()
```

This code is creating a new column called **"Review_Category"** in the DataFrame. It first initializes all values in this column to 'No Reviews mention'. Then, based on the values in the 'Reviews' column, it assigns different categories to the apps' review counts. For example, if an app has reviews between 1 and 1000, it's categorized as 'Less than 1K Reviews'. The same logic applies for other categories like 'Between 1K and 10K Reviews', 'Between 10K and 100K Reviews', and so on. Finally, it counts the number of apps in each review category using the **value_counts()** function. This code essentially categorizes apps based on the number of reviews they have received.

```python
df1_user_rating = pd.crosstab(df["Primary_Genre"], df["User_Rating"]).apply(lambda x: round(x/x.sum()*100,1), axis=1).sort_values(by="No Rating", ascending=False)
df2_type_user_rating = pd.crosstab(df["Primary_Genre"], df["Type_User_Rating"]).apply(lambda x: round(x/x.sum()*100,1), axis=1).sort_values(by=["Rating 4 to 5"], ascending=False)
df3_review_category = pd.crosstab(df["Primary_Genre"], df["Review_Category"]).apply(lambda x: round(x/x.sum()*100,2), axis=1).sort_values(by="More than 1M Reviews", ascending=False)
df4_free_paid = pd.crosstab(df["Primary_Genre"], df["Type"]).apply(lambda x: round(x/x.sum()*100,2), axis=1).sort_values(by="Free", ascending=False)
df5_rating = pd.merge(df1_user_rating, df2_type_user_rating, on="Primary_Genre", how="outer")
df6_review_free_paid = pd.merge(df3_review_category, df4_free_paid, on="Primary_Genre", how="outer")
df7_rating_free_paid_review = pd.merge(df5_rating, df6_review_free_paid, on="Primary_Genre", how="outer").drop(["No Rating_y", "No Reviews mention"], axis=1).rename(columns={'No Rating_x' : 'No Rating'})
df8_rating_free_paid_review = df7_rating_free_paid_review.iloc[::,[0,1,2,3,4,5,6,13,14,11,9,8,7,10,12]]
df8_rating_free_paid_review
```

**df2_type_user_rating:** Computing the percentage distribution of user rating types (No Rating, Rating 0 to 1, etc.) for each primary genre.

**df3_review_category:** Generating the percentage distribution of review categories (Less than 1K Reviews, Between 1K and 10K Reviews, etc.) for each primary genre.

**df4_free_paid:** Creating the percentage distribution of app types (Free or Paid) for each primary genre.

Merging the dataframes **df1_user_rating** and **df2_type_user_rating** into **df5_rating**.

Merging the dataframes **df3_review_category** and **df4_free_paid** into **df6_review_free_paid**.

Merging **df5_rating** and **df6_review_free_paid** into **df7_rating_free_paid_review**. Columns that are not needed are dropped and columns are renamed appropriately.

Rearranging the columns for better readability in **df8_rating_free_paid_review**.

The final result, **df8_rating_free_paid_review**, is a DataFrame that provides insights into the distribution of user ratings, user rating types, review categories, and app types (Free or Paid) across different primary genres in the dataset.
